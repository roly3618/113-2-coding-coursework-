```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets_2 = gc.open_by_url('https://docs.google.com/spreadsheets/d/1fpCKNCE1HnCpdf4Exy7--6feYAI1w95YT3yJce8zo7w/edit?usp=sharing')

gsheets_2
```

```python
# 取得 Google Sheets 內容(藥品)
data_2 = worksheet_2.get_all_values()  # 取得所有儲存格資料，回傳為 2D 陣列
df_2 = pd.DataFrame(data_2)
```

```python
from datetime import datetime, timedelta
```

```python
# 建立 06:00 ~ 23:00 的每小時時間點
time_slots = [datetime.strptime("06:00", "%H:%M") + timedelta(hours=i) for i in range(19)]  # 6:00 到 24:00

# 建立時段字典
schedule_by_hour = {t.strftime("%H:00"): [] for t in time_slots}

df_2 = pd.DataFrame(data_2[1:], columns=data_2[0])  # 使用第0列當欄位名稱

# 幫每個藥分配時間點
for index, row in df_2.iterrows():
    drug_name = row["藥品名稱"]
    times_per_day = int(row["使用間隔(次/每日)"])

    if times_per_day == 1:
        chosen_slots = [time_slots[len(time_slots) // 2]]
    else:
        interval = len(time_slots) / times_per_day
        chosen_slots = [
            time_slots[min(int(i * interval + interval / 2), len(time_slots) - 1)]
            for i in range(times_per_day)
        ]

    for slot in chosen_slots:
        key = slot.strftime("%H:00")
        schedule_by_hour[key].append(drug_name)

# 轉成 DataFrame
hourly_schedule = pd.DataFrame([
    {"時間": hour, "藥品清單/所需照護": ", ".join(drugs) if drugs else ""}
    for hour, drugs in schedule_by_hour.items()
])

hourly_schedule
```

```python
# ✅ Google Colab 整合教學：照護排程結果匯出至 Google Sheets（動態讀取兩份試算表，含病人姓名）

# ---
# 第一步：安裝必要套件（Colab 通常已內建，但可保險）
!pip install --upgrade gspread gspread-dataframe oauth2client

# ---
# 第二步：使用 Colab OAuth 授權免金鑰登入
from google.colab import auth
auth.authenticate_user()

import gspread
import pandas as pd
from gspread_dataframe import set_with_dataframe
from google.auth import default
creds, _ = default()
client = gspread.authorize(creds)

# ---
# 第三步：分別開啟兩份 Google Sheet
sheet_drugs = client.open_by_key("1M8asf8VKqY8Fv3L4ZHM9vXG1Dsr-l3A-eZ-4As80ON4").worksheet("工作表1")
sheet_patients = client.open_by_key("1qTxmc63YYxn6qR09dDm51OrcY4AG0AeYccCcv9Pz2Qg").worksheet("工作表1")

df_drugs = pd.DataFrame(sheet_drugs.get_all_records())
df_patients = pd.DataFrame(sheet_patients.get_all_records())

# ---
# 第四步：建立每日照護排程（06:00~23:00）
from datetime import datetime, timedelta

time_slots = [datetime.strptime("06:00", "%H:%M") + timedelta(hours=i) for i in range(18)]
schedule_records = []

# 建立藥品時間排程參考表
drug_time_map = {}
for index, row in df_drugs.iterrows():
    name = row["藥品名稱"]
    times_per_day = int(row["使用間隔(次/每日)"])

    if times_per_day == 1:
        chosen_slots = [time_slots[len(time_slots) // 2]]
    else:
        interval = len(time_slots) / times_per_day
        chosen_slots = [
            time_slots[min(int(i * interval + interval / 2), len(time_slots) - 1)]
            for i in range(times_per_day)
        ]
    drug_time_map[name] = [slot.strftime("%H:00") for slot in chosen_slots]

# 依病人需求對照藥品時間，加上病人姓名與樓層等資訊
for _, patient in df_patients.iterrows():
    name = patient["病人姓名"]
    floor = patient["所在樓層"]
    level = patient.get("照護等級", "")
    emergency = patient.get("症狀緊急程度", "")

    for field in ["需要的藥物1", "需要的藥物2", "照護"]:
        item = patient.get(field)
        if item and item in drug_time_map:
            for time in drug_time_map[item]:
                schedule_records.append({
                    "時間": time,
                    "病人姓名": name,
                    "樓層": floor,
                    "照護/藥品項目": item,
                    "照護等級": level,
                    "緊急程度": emergency
                })

# 建立 DataFrame 並排序
hourly_schedule = pd.DataFrame(schedule_records)
hourly_schedule = hourly_schedule.sort_values(by=["時間", "樓層", "照護等級", "緊急程度"], ascending=[True, True, False, False])

# ---
# 第五步：寫入新的 Google Sheets 分頁（你可以寫回任一 Sheet）
# 這裡以病人那份為輸出對象
spreadsheet_out = client.open_by_key("1ctVg_q70Ljeoltqq61Zn9cgG5dsMYaAWvhGREjy3Ug0")
worksheet_name = "照護排程"
try:
    worksheet = spreadsheet_out.worksheet(worksheet_name)
    worksheet.clear()  # 避免刪除重建，降低 API 限制風險
except:
    worksheet = spreadsheet_out.add_worksheet(title=worksheet_name, rows="200", cols="20")

set_with_dataframe(worksheet, hourly_schedule)
print("✅ 排程已成功匯出至 Google Sheets『照護排程』分頁（含病人姓名）！")
```
